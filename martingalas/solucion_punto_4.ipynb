**Escenario A:** Supón que cuentas con 500 mil pesos para apostar y se parará el juego cuando te quedes sin dinero o cuando apuestes 50 veces (de acuerdo a las políticas del casino).

```python
import numpy as np
import matplotlib.pyplot as plt

def Simul_martingala(capital_inicial, max_apuestas):
    capital_total = capital_inicial
    perdida = 0
    max_apuestas = 50
    monto_apuesta = 10000
    apuestas = []
    cont_apuesta = 0
    while (capital_total > 2*monto_apuesta and cont_apuesta < max_apuestas):
        apuestas.append(capital_total)
        cont_apuesta += 1
        negro = np.random.choice([True, False])
        if perdida == 0:
            monto_apuesta = 10000
        else:
            monto_apuesta = 2 * perdida
        if negro:
            capital_total += monto_apuesta
            perdida = 0
        else:
            capital_total = capital_total - monto_apuesta
            perdida += monto_apuesta


    cont_apuesta += 1

    apuestas.append(0)
    return capital_total, cont_apuesta-1


ganancia_final = []
napuestas = []
N_simulaciones = 500

for i in range(N_simulaciones):
    x, y = Simul_martingala(500000, 50)
    ganancia_final.append(x)
    napuestas.append(y)

promedio_dinero = np.mean(ganancia_final)
promedio_num_apuestas = np.mean(napuestas)

varianza_dinero = np.std(ganancia_final)
varianza_num_apuestas = np.std(napuestas)

max_dinero_obtenido = np.max(ganancia_final)
max_num_apuestas = np.max(napuestas)

min_dinero = np.min(ganancia_final)
min_num_apuestas = np.min(napuestas)

print("Después de 500 simulaciones:")
print(f"- El promedio de ganancias obtenidas es: {promedio_dinero}")
print(f"- El promedio de apuestas realizadas es: {round(promedio_num_apuestas, 2)}")
print(f"- El máximo de ganancias obtenidas es: {max_dinero_obtenido}")
print(f"- El máximo número de apuestas realizadas es: {max_num_apuestas}")
print(f"- El mínimo de ganancias obtenidas es: {min_dinero}")
print(f"- El mínimo número de apuestas realizadas es: {min_num_apuestas}")
print(f"- La desviación estándar de las ganancias obtenidas es: {round(varianza_dinero, 2)}")
print(f"- La desviación estándar del número de apuestas realizadas es: {round(varianza_num_apuestas, 2)}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
ax1.hist(ganancia_final)
ax1.set_title("Frecuencia Ganancias finales")
ax1.set_xlabel("Ganancia Final")
ax1.set_ylabel("Frecuencia")
ax2.hist(napuestas)
ax2.set_title("Frecuencia Número de apuestas")
ax2.set_xlabel("Número de Apuestas")
ax2.set_ylabel("Frecuencia")
plt.show()
```
